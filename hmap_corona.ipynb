```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_excel('확진자_국민행복지수상관계수.xlsx')
## 한글 폰트 깨짐 때문에 추가 ################# 주피터일때
#from matplotlib import font_manager, rc  

#plt.rcParams['axes.unicode_minus'] = False
#f_path = 'c:/Windows/Fonts/malgun.ttf'
#font_name= font_manager.FontProperties(fname=f_path).get_name()
#rc('font', family=font_name)
#################################################
## 코랩일때 한글폰트 ############################
!apt-get update -qq
!apt-get install fonts-nanum* -qq
PATH = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
from matplotlib import font_manager, rc
import matplotlib.font_manager as fm
plt.rcParams['axes.unicode_minus'] = False
f_path = PATH
font_name= font_manager.FontProperties(fname=f_path).get_name()
plt.rc('font', family=font_name)
fm._rebuild()
##################################################

df.rename(columns={'Unnamed: 0':'자치구'}, inplace=True)
df.set_index(keys='자치구', inplace=True)     # 인덱스 재지정


plt.figure(figsize=(20,5))
plt.xticks(rotation=45)
sns.heatmap(df.T, annot=True, cmap='coolwarm', vmin= -1, vmax= 1)   ## df.T해서 가로형으로 바꿔줌. 컬러바의 값은 -1 ~ 1까지(상관계수니까)
plt.title('확진자 수와 행복지수(자치구별)')

plt.show()
```
