```python
#함수 /클래스 선언부
class Node():
    def __init__(self):
        self.data = None
        self.link = None
        
def printNode(start):
    current = start
    print(current.data, end =' ')
    while current.link != None:
        current = current.link
        print(current.data, end=' ')
    print(' ')

def insertNode(findData, insertData):  #find 앞에 insert를 넣는다.
    global memory, head, current, pre 
    #첫노드 앞에 삽입하는 경우 handling
    if head.data == findData :   
        node = Node()
        node.data = insertData
        node.link = head
        head = node  
        memory.append(node)
        return
# 중간에 삽입하기. 
    current = head #시작
    while current.link != None :   # 끝노드가 아니면 
        pre = current            # 전노드를 잡고서 (어디삽입할지 위치알아야)
        current = current.link  # 하나 옮겨간다
        if current.data == findData :
            node = Node()
            node.data= insertData
            node.link = current
            pre.link = node 
            memory.append(node)
            return
# 찾는 데이터 없으면 맨뒤에 삽입
    node= Node()
    node.data = insertData
    current.link = node
    memory.append(node)

def deleteNode(deleteData):
    global memory, head, current, pre 
    # 1st node delete
    if deleteData == head.data:
        current = head
        head = head.link
        del(current)
        return
    # other nodes delete
    current = head 
    while current.link != None : 
        pre = current
        current = current.link
        if current.data == deleteData : 
            pre.link = current.link
            del(current)
            return

def findNode(findData):
    global memory, head, pre, current 
    current = head 
    if current.data == findData:
        return current
    while current.link != None:
        current = current.link
        if current.data == findData:
            return current
    return Node()
#전역변수
memory = []  #not necessary, but for constancy with othr langs.
head, current, pre = None, None, None
dataArray = ['a','b','c','d','e']
#메인 코드부 
node = Node()   #재사용할 변수 node
node.data=dataArray[0]
head = node
memory.append(node)  #이코드는 안써도 무방.. 

for data in dataArray[1:]:  #첫번째 요소 이후 나머지 항목에서 돔.
    pre = node #다음꺼 만들기 전에 현재노드 이전 데이터 잡고있기.
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node)
    
printNode(head)
    
insertNode('a','new')
printNode(head)
insertNode('c', 'mid')
printNode(head)
insertNode('z', 'tail')
printNode(head)
deleteNode('a')
printNode(head)
deleteNode('c')
printNode(head)
deleteNode('l')
printNode(head)

fNode=findNode('b')
print(fNode.data)
fNode=findNode('mid')
print(fNode.data)
fNode = findNode('x')
print(fNode.data)
```
