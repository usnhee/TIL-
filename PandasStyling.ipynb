# Pandas Styling

```python
import pandas as pd
import numpy as np

np.random.seed(88)
df = pd.DataFrame({'A': np.linspace(1,10,10)})
df = pd.concat([df, pd.DataFrame(np.random.randn(10,4), columns=list('BCDE'))], axis=1)
df.iloc[3,3] = np.nan
df.iloc[0,2] = np.nan

print(df)
```

```python
df.style.highlight_max()
```

```python
df.style.highlight_max(axis=1)
```

```python
df.style.highlight_min()
```

```python
df.style.highlight_null()
```

```python
df.style.set_na_rep('OutofScope').highlight_null(null_color='orange')
```

```python
df.style.background_gradient()    # heatmap
```

1. **cmap** : by default, the 'PuBu' colormap is selected by pandas. 
You can create a custom matplotlib colormap and pass it to 
the cmap parameter.
2. **axis** : generating heat via rows or columns criteria, by default: 
columns
3. **text_color_threshold** : controls text visibility across varying 
background colors

```python
df.style.set_properties(**{'border':'1.3px solid green', 'color': 'magenta'})
```

```python
df.style.bar(color = 'green')
```

```python
df.style.set_precision(2)   # 소수점 자리수 지정
```

```python
df.style.set_caption("This is DataFrame styling demo.").set_precision(2).background_gradient()
```

```python
df.style.hide_index()
```

```python
df.style.hide_columns('B')
```

```python
df.style.format("{:.3%}")   
## 소수점아래 셋째자리까지 표시, 퍼센트로.
```

```python
df.style.format("{:.3%}", na_rep="$$")   
## na 표시는 $$로 
```

```python
df.style.set_precision(2).background_gradient().hide_index().to_excel('styled.xlsx', engine='openpyxl')
```
