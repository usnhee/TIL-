```python
def find_places(searching):
    # 접속 url 만들기
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(searching)
    
    # headers 입력
    headers = { "Authorization": "KakaoAK ee2a3173db5d2791e754200d817ef1f9"}
    
    #API 요청 및 정보 받기 
    places=requests.get(url, headers=headers).json()['documents']
    
    # 필요 정보 선택 
    place = places[0]    # 1st value saved in 'place'
    name = place['place_name']   # save the name of place
    x = place['x']  # bring x, y values.  # latitude
    y = place['y']                        #longitude   
    data = [name, x, y, searching]    # save as list 
    
    return data 
```

```python
from folium.plugins import MarkerCluster

locations = []
names = []
Mt_Hala = [33.362500, 126.533694]     # 센터로 한라산 찍어놓기 
map_jeju2 = folium.Map(location=Mt_Hala, zoom_start=11)

for i in range(len(location_data)):
    data = location_data.iloc[i]   # 행 하나씩 
    locations.append((float(data['latitude']), float(data['longitude']))) #위도 경도 순  
    names.append(data['name_official'])
    
marker_cluster = MarkerCluster(locations=locations, popups=names, name='Jeju', overlay=True, control=True)

marker_cluster.add_to(map_jeju2)
folium.LayerControl().add_to(map_jeju2)

tiles = ['stamenwatercolor','cartodbpositron', 'openstreetmap']
for tile in tiles:
    folium.TileLayer(tile).add_to(map_jeju2)

map_jeju2
```
